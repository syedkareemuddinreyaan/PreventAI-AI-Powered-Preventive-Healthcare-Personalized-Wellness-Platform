# PreventAI – Final Integration Demo

## Purpose

This notebook integrates all PreventAI modules:

- Health Score Engine
- Workout Recommendation Engine
- Diet Recommendation Engine
- Health Improvement Simulator

It demonstrates the end-to-end flow of the system using sample data.

```python
prediction_data = {
    "heart_risk": "Medium",
    "diabetes_risk": "Low",
    "obesity_status": "Obese",
    "sleep_risk": "Poor",
    "stress_risk": "Moderate"
}

user_data = {
    "age": 21,
    "gender": "Male",
    "weight": 90,
    "height": 175,
    "activity_level": "Sedentary",
    "goal": "Weight Loss"
}
```

```python
heart_scores = {
    "Low": 100,
    "Medium": 70,
    "High": 40
}

diabetes_scores = {
    "Low": 100,
    "Medium": 70,
    "High": 40
}

obesity_scores = {
    "Underweight": 70,
    "Healthy": 100,
    "Overweight": 70,
    "Obese": 40
}

sleep_scores = {
    "Good": 100,
    "Moderate": 70,
    "Poor": 40
}

stress_scores = {
    "Low": 100,
    "Moderate": 70,
    "High": 40
}
```

```python
def calculate_health_score(predictions):

    # Convert risks into numerical scores
    heart = heart_scores[predictions["heart_risk"]]

    diabetes = diabetes_scores[predictions["diabetes_risk"]]

    obesity = obesity_scores[predictions["obesity_status"]]

    sleep = sleep_scores[predictions["sleep_risk"]]

    stress = stress_scores[predictions["stress_risk"]]


    # Apply weights
    final_score = (
        heart * 0.30 +
        diabetes * 0.20 +
        obesity * 0.20 +
        sleep * 0.15 +
        stress * 0.15
    )

    return round(final_score)
```

```python
def health_category(score):

    if score >= 85:
        return "Excellent"

    elif score >= 70:
        return "Good"

    elif score >= 50:
        return "Moderate"

    else:
        return "Poor"
```

```python
score = calculate_health_score(prediction_data)

category = health_category(score)

print("===== HEALTH REPORT =====")

print("Health Score:", score)

print("Category:", category)

print("=========================")
```

```python
def recommend_workout(predictions, user):

    goal = user["goal"]
    heart_risk = predictions["heart_risk"]
    obesity_status = predictions["obesity_status"]

    if goal == "Weight Loss":

        workout = {
            "Training Type": "Cardio + Resistance Training",

            "Weekly Plan": [
                "Walking - 30 minutes (5 days/week)",
                "Cycling - 20 minutes (3 days/week)",
                "Strength Training - 2 days/week"
            ],

            "Avoid": []
        }

    elif goal == "Muscle Gain":

        workout = {
            "Training Type": "Strength Focused Training",

            "Weekly Plan": [
                "Upper Body - 2 days/week",
                "Lower Body - 2 days/week",
                "Core Training - 1 day/week",
                "Light Cardio - 2 days/week"
            ],

            "Avoid": []
        }

    else:

        workout = {
            "Training Type": "Balanced Fitness",

            "Weekly Plan": [
                "Walking - 30 minutes",
                "Strength Training - 2 days/week",
                "Stretching - Daily"
            ],

            "Avoid": []
        }

    if heart_risk == "High":

        workout["Avoid"].extend([
            "HIIT",
            "Heavy Weight Lifting"
        ])

        workout["Weekly Plan"] = [
            "Walking - 30 minutes",
            "Cycling - 20 minutes",
            "Swimming - 20 minutes"
        ]

    if obesity_status == "Obese":

        workout["Weekly Plan"].append(
            "Increase Daily Steps to 8000–10000"
        )

    return workout
```

```python
workout_plan = recommend_workout(
    prediction_data,
    user_data
)
```

```python
print("\n===== WORKOUT PLAN =====")

print("Training Type:")
print(workout_plan["Training Type"])

print("\nWeekly Plan:")

for activity in workout_plan["Weekly Plan"]:
    print("-", activity)

print("\nAvoid:")

if workout_plan["Avoid"]:
    for item in workout_plan["Avoid"]:
        print("-", item)
else:
    print("None")

print("========================")
```

```python
def calculate_bmi(weight, height):
    
    height_m = height / 100
    
    bmi = weight / (height_m ** 2)
    
    return round(bmi, 2)
```

```python
def bmi_category(bmi):
    
    if bmi < 18.5:
        return "Underweight"
    
    elif bmi < 25:
        return "Healthy"
    
    elif bmi < 30:
        return "Overweight"
    
    else:
        return "Obese"
```

```python
def calculate_bmr(weight, height, age, gender):
    
    if gender == "Male":
        
        bmr = (
            10 * weight +
            6.25 * height -
            5 * age +
            5
        )
    
    else:
        
        bmr = (
            10 * weight +
            6.25 * height -
            5 * age -
            161
        )
    
    return round(bmr)
```

```python
def calculate_tdee(bmr, activity_level):
    
    multipliers = {
        "Sedentary": 1.2,
        "Light": 1.375,
        "Moderate": 1.55,
        "Active": 1.725
    }
    
    tdee = bmr * multipliers[activity_level]
    
    return round(tdee)
```

```python
def calculate_goal_calories(tdee, goal):

    if goal == "Weight Loss":
        calories = tdee - 500

    elif goal == "Weight Gain":
        calories = tdee + 300

    elif goal == "Muscle Gain":
        calories = tdee + 200

    else:
        calories = tdee

    return round(calories)
```

```python
def calculate_protein(weight, goal):

    if goal == "Weight Loss":
        protein = weight * 1.8

    elif goal == "Muscle Gain":
        protein = weight * 2.0

    elif goal == "Weight Gain":
        protein = weight * 1.6

    else:
        protein = weight * 1.5

    return round(protein)
```

```python
def calculate_fats(calories):

    fat_calories = calories * 0.25

    fats = fat_calories / 9

    return round(fats)
```

```python
def calculate_carbs(calories, protein, fats):

    protein_calories = protein * 4

    fat_calories = fats * 9

    remaining_calories = (
        calories -
        protein_calories -
        fat_calories
    )

    carbs = remaining_calories / 4

    return round(carbs)
```

```python
meal_plans = {

    "Weight Loss": {

        "Breakfast": [
            "Oats",
            "Boiled Eggs",
            "Fruit"
        ],

        "Lunch": [
            "Brown Rice",
            "Grilled Chicken",
            "Vegetables"
        ],

        "Dinner": [
            "Paneer",
            "Salad"
        ]
    },

    "Muscle Gain": {

        "Breakfast": [
            "Oats",
            "Peanut Butter",
            "Banana"
        ],

        "Lunch": [
            "Rice",
            "Chicken",
            "Vegetables"
        ],

        "Dinner": [
            "Paneer",
            "Rice"
        ]
    },

    "Weight Gain": {

        "Breakfast": [
            "Eggs",
            "Milk",
            "Banana"
        ],

        "Lunch": [
            "Rice",
            "Chicken",
            "Potatoes"
        ],

        "Dinner": [
            "Paneer",
            "Chapati"
        ]
    },

    "General Fitness": {

        "Breakfast": [
            "Oats",
            "Fruit"
        ],

        "Lunch": [
            "Rice",
            "Dal",
            "Vegetables"
        ],

        "Dinner": [
            "Salad",
            "Paneer"
        ]
    }
}
```

```python
bmi = calculate_bmi(
    user_data["weight"],
    user_data["height"]
)

bmr = calculate_bmr(
    user_data["weight"],
    user_data["height"],
    user_data["age"],
    user_data["gender"]
)

tdee = calculate_tdee(
    bmr,
    user_data["activity_level"]
)

goal_calories = calculate_goal_calories(
    tdee,
    user_data["goal"]
)

protein = calculate_protein(
    user_data["weight"],
    user_data["goal"]
)

fats = calculate_fats(goal_calories)

carbs = calculate_carbs(
    goal_calories,
    protein,
    fats
)
```

```python
print("\n===== DIET REPORT =====")

print("BMI:", bmi)

print("BMI Category:", bmi_category(bmi))

print("Daily Calories:", goal_calories)

print("Protein:", protein, "g")

print("Fats:", fats, "g")

print("Carbohydrates:", carbs, "g")

print("\nMeal Plan:")

for meal, foods in meal_plans[user_data["goal"]].items():

    print(f"\n{meal}:")

    for food in foods:
        print("-", food)

print("=======================")
```

```python
def simulate_future_health(
    current_score,
    improved_sleep,
    regular_exercise,
    improved_bmi
):

    future_score = current_score

    if improved_sleep:
        future_score += 10

    if regular_exercise:
        future_score += 10

    if improved_bmi:
        future_score += 15

    future_score = min(future_score, 100)

    return future_score
```

```python
future_score = simulate_future_health(
    current_score=score,
    improved_sleep=True,
    regular_exercise=True,
    improved_bmi=True
)

print("\n===== FUTURE HEALTH =====")

print("Current Health Score:", score)

print("Projected Health Score:", future_score)

print("=========================")
```
