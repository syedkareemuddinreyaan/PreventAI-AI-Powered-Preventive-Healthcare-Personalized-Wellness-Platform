```python
def simulate_future_health(
    current_score,
    improved_sleep,
    regular_exercise,
    improved_bmi
):

    future_score = current_score

    if improved_sleep:
        future_score += 10

    if regular_exercise:
        future_score += 10

    if improved_bmi:
        future_score += 15

    future_score = min(future_score, 100)

    return future_score
```

```python
future_score = simulate_future_health(
    current_score=current_health_score,
    improved_sleep=True,
    regular_exercise=True,
    improved_bmi=True
)

print("Future Health Score:", future_score)
```

```python
print("===== HEALTH IMPROVEMENT REPORT =====")

print("Current Health Score:", current_health_score)

print("\nLifestyle Changes:")

print("- Improved Sleep: Yes")

print("- Regular Exercise: Yes")

print("- Improved BMI: Yes")

print("\nProjected Health Score:", future_score)

print("====================================")
```

```python
future_score_2 = simulate_future_health(
    current_score=current_health_score,
    improved_sleep=True,
    regular_exercise=False,
    improved_bmi=False
)

print("Projected Score:", future_score_2)
```
