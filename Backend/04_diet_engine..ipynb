# PreventAI – Diet Recommendation Engine

## Purpose

This notebook generates personalized diet recommendations based on:

- Age
- Gender
- Weight
- Height
- Activity Level
- Health Goal

The engine calculates:
- BMI
- BMR
- TDEE
- Daily Calories
- Macronutrient Requirements
- Meal Suggestions

```python
user_data = {
    "age": 21,
    "gender": "Male",
    "weight": 90,      # kg
    "height": 175,     # cm
    "activity_level": "Sedentary",
    "goal": "Weight Loss"
}
```

```python
def calculate_bmi(weight, height):
    
    height_m = height / 100
    
    bmi = weight / (height_m ** 2)
    
    return round(bmi, 2)
```

```python
bmi = calculate_bmi(
    user_data["weight"],
    user_data["height"]
)

print("BMI:", bmi)
```

```python
def bmi_category(bmi):
    
    if bmi < 18.5:
        return "Underweight"
    
    elif bmi < 25:
        return "Healthy"
    
    elif bmi < 30:
        return "Overweight"
    
    else:
        return "Obese"
```

```python
print("BMI Category:", bmi_category(bmi))
```

```python
def calculate_bmr(weight, height, age, gender):
    
    if gender == "Male":
        
        bmr = (
            10 * weight +
            6.25 * height -
            5 * age +
            5
        )
    
    else:
        
        bmr = (
            10 * weight +
            6.25 * height -
            5 * age -
            161
        )
    
    return round(bmr)
```

```python
bmr = calculate_bmr(
    user_data["weight"],
    user_data["height"],
    user_data["age"],
    user_data["gender"]
)

print("BMR:", bmr)
```

```python
def calculate_tdee(bmr, activity_level):
    
    multipliers = {
        "Sedentary": 1.2,
        "Light": 1.375,
        "Moderate": 1.55,
        "Active": 1.725
    }
    
    tdee = bmr * multipliers[activity_level]
    
    return round(tdee)
```

```python
tdee = calculate_tdee(
    bmr,
    user_data["activity_level"]
)

print("TDEE:", tdee)
```

```python
def calculate_goal_calories(tdee, goal):

    if goal == "Weight Loss":
        calories = tdee - 500

    elif goal == "Weight Gain":
        calories = tdee + 300

    elif goal == "Muscle Gain":
        calories = tdee + 200

    else:
        calories = tdee

    return round(calories)
```

```python
goal_calories = calculate_goal_calories(
    tdee,
    user_data["goal"]
)

print("Daily Calorie Target:", goal_calories)
```

```python
def calculate_protein(weight, goal):

    if goal == "Weight Loss":
        protein = weight * 1.8

    elif goal == "Muscle Gain":
        protein = weight * 2.0

    elif goal == "Weight Gain":
        protein = weight * 1.6

    else:
        protein = weight * 1.5

    return round(protein)
```

```python
protein = calculate_protein(
    user_data["weight"],
    user_data["goal"]
)

print("Protein Requirement:", protein, "g")
```

```python
def calculate_fats(calories):

    fat_calories = calories * 0.25

    fats = fat_calories / 9

    return round(fats)
```

```python
fats = calculate_fats(goal_calories)

print("Fat Requirement:", fats, "g")
```

```python
def calculate_carbs(calories, protein, fats):

    protein_calories = protein * 4

    fat_calories = fats * 9

    remaining_calories = (
        calories -
        protein_calories -
        fat_calories
    )

    carbs = remaining_calories / 4

    return round(carbs)
```

```python
carbs = calculate_carbs(
    goal_calories,
    protein,
    fats
)

print("Carbohydrate Requirement:", carbs, "g")
```

```python
meal_plans = {

    "Weight Loss": {

        "Breakfast": [
            "Oats",
            "Boiled Eggs",
            "Fruit"
        ],

        "Lunch": [
            "Brown Rice",
            "Grilled Chicken",
            "Vegetables"
        ],

        "Dinner": [
            "Paneer",
            "Salad"
        ]
    },

    "Muscle Gain": {

        "Breakfast": [
            "Oats",
            "Peanut Butter",
            "Banana"
        ],

        "Lunch": [
            "Rice",
            "Chicken",
            "Vegetables"
        ],

        "Dinner": [
            "Paneer",
            "Rice"
        ]
    },

    "Weight Gain": {

        "Breakfast": [
            "Eggs",
            "Milk",
            "Banana"
        ],

        "Lunch": [
            "Rice",
            "Chicken",
            "Potatoes"
        ],

        "Dinner": [
            "Paneer",
            "Chapati"
        ]
    },

    "General Fitness": {

        "Breakfast": [
            "Oats",
            "Fruit"
        ],

        "Lunch": [
            "Rice",
            "Dal",
            "Vegetables"
        ],

        "Dinner": [
            "Salad",
            "Paneer"
        ]
    }
}
```

```python
goal = user_data["goal"]

print("========== DIET REPORT ==========")

print("BMI:", bmi)

print("BMI Category:", bmi_category(bmi))

print("Daily Calories:", goal_calories)

print("Protein:", protein, "g")

print("Fats:", fats, "g")

print("Carbohydrates:", carbs, "g")

print("\nMeal Plan:")

for meal, foods in meal_plans[goal].items():

    print(f"\n{meal}:")

    for food in foods:
        print("-", food)

print("\n=================================")
```
