# PreventAI – Workout Recommendation Engine

## Purpose
This notebook generates personalized workout recommendations based on:

- User fitness goals
- Heart disease risk
- Obesity status

The system aims to provide safe and effective exercise plans tailored to individual needs.

```python
prediction_data = {
    "heart_risk": "Medium",
    "diabetes_risk": "Low",
    "obesity_status": "Obese",
    "sleep_risk": "Poor",
    "stress_risk": "Moderate"
}

user_data = {
    "age": 21,
    "gender": "Male",
    "weight": 90,
    "height": 175,
    "activity_level": "Sedentary",
    "goal": "Weight Loss"
}
```

```python
def recommend_workout(predictions, user):

    goal = user["goal"]
    heart_risk = predictions["heart_risk"]
    obesity_status = predictions["obesity_status"]

    # Goal-based recommendations
    if goal == "Weight Loss":

        workout = {
            "Training Type": "Cardio + Resistance Training",

            "Weekly Plan": [
                "Walking - 30 minutes (5 days/week)",
                "Cycling - 20 minutes (3 days/week)",
                "Strength Training - 2 days/week"
            ],

            "Avoid": []
        }

    elif goal == "Muscle Gain":

        workout = {
            "Training Type": "Strength Focused Training",

            "Weekly Plan": [
                "Upper Body - 2 days/week",
                "Lower Body - 2 days/week",
                "Core Training - 1 day/week",
                "Light Cardio - 2 days/week"
            ],

            "Avoid": []
        }

    else:

        workout = {
            "Training Type": "Balanced Fitness",

            "Weekly Plan": [
                "Walking - 30 minutes",
                "Strength Training - 2 days/week",
                "Stretching - Daily"
            ],

            "Avoid": []
        }

    # Safety rule for high heart risk
    if heart_risk == "High":

        workout["Avoid"].extend([
            "HIIT",
            "Heavy Weight Lifting"
        ])

        workout["Weekly Plan"] = [
            "Walking - 30 minutes",
            "Cycling - 20 minutes",
            "Swimming - 20 minutes"
        ]

    # Extra recommendation for obesity
    if obesity_status == "Obese":

        workout["Weekly Plan"].append(
            "Increase Daily Steps to 8000–10000"
        )

    return workout
```

```python
workout_plan = recommend_workout(
    prediction_data,
    user_data
)

print(workout_plan)
```
