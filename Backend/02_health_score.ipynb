# PreventAI – Health Score Engine

## Purpose
This notebook generates an overall Health Score (0–100) based on:

- Heart Disease Risk
- Diabetes Risk
- Obesity Status
- Sleep Health Risk
- Stress Risk

The Health Score helps users quickly understand their overall health condition.

```python
prediction_data = {
    "heart_risk": "Medium",
    "diabetes_risk": "Low",
    "obesity_status": "Obese",
    "sleep_risk": "Poor",
    "stress_risk": "Moderate"
}
```

```python
# Heart Disease Risk Scores
heart_scores = {
    "Low": 100,
    "Medium": 70,
    "High": 40
}

# Diabetes Risk Scores
diabetes_scores = {
    "Low": 100,
    "Medium": 70,
    "High": 40
}

# Obesity Status Scores
obesity_scores = {
    "Underweight": 70,
    "Healthy": 100,
    "Overweight": 70,
    "Obese": 40
}

# Sleep Risk Scores
sleep_scores = {
    "Good": 100,
    "Moderate": 70,
    "Poor": 40
}

# Stress Risk Scores
stress_scores = {
    "Low": 100,
    "Moderate": 70,
    "High": 40
}
```

```python
def calculate_health_score(predictions):

    # Convert risks into numerical scores
    heart = heart_scores[predictions["heart_risk"]]

    diabetes = diabetes_scores[predictions["diabetes_risk"]]

    obesity = obesity_scores[predictions["obesity_status"]]

    sleep = sleep_scores[predictions["sleep_risk"]]

    stress = stress_scores[predictions["stress_risk"]]


    # Apply weights
    final_score = (
        heart * 0.30 +
        diabetes * 0.20 +
        obesity * 0.20 +
        sleep * 0.15 +
        stress * 0.15
    )

    return round(final_score)
```

```python
score = calculate_health_score(prediction_data)

print("Health Score:", score)
```

```python
def health_category(score):

    if score >= 85:
        return "Excellent"

    elif score >= 70:
        return "Good"

    elif score >= 50:
        return "Moderate"

    else:
        return "Poor"
```

```python
category = health_category(score)

print("Health Category:", category)
```

```python
print("========== HEALTH REPORT ==========")

print("Health Score:", score, "/100")

print("Health Category:", category)

print("===================================")
```
